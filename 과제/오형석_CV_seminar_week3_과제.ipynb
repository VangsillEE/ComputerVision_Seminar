```python
import matplotlib.pyplot as plt
import glob
from google.colab.patches import cv2_imshow
import cv2
from PIL import Image as PILImage
import numpy as np
```

```python
dolphin_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/test/dolphin/*')
shark_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/test/shark/*')
whale_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/test/whale/*')
```

```python
#1
fig = plt.figure(figsize=(10,10))
axs = fig.subplots(3,3).flatten()
axs[0].set_title("dolpin")
axs[1].set_title("shark")
axs[2].set_title("whale")
for num in range(1,10):
  if num in [1,4,7]:
    image = cv2.imread(dolphin_img_list[num])
  elif num in [2,5,8]:     
    image = cv2.imread(shark_img_list[num])
  elif num in [3,6,9]:
    image = cv2.imread(whale_img_list[num])
  axs[num-1].imshow(image[:,:,::-1])
  axs[num-1].set_xlabel(image.shape)
```

```python
#2
image_1 = cv2.imread(dolphin_img_list[10]) 
num6 = int(image_1.shape[0]/6)
image_1[num6:num6+10,15:-15,] =np.array([0,0,0]*10*153).reshape(10,153,3)
image_1[2*num6:2*num6+10,15:-15,] =np.array([0,0,255]*10*153).reshape(10,153,3)
image_1[3*num6:3*num6+10,15:-15,] =np.array([0,255,0]*10*153).reshape(10,153,3)
image_1[4*num6:4*num6+10,15:-15,] =np.array([255,0,0]*10*153).reshape(10,153,3)
image_1[5*num6:5*num6+10,15:-15,] =np.array([255,255,255]*10*153).reshape(10,153,3)
cv2_imshow(image_1)
```
